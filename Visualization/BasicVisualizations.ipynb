```python
# Libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Line:
x = np.arange(1,11)
y = np.random.randint(1,100, 10)

plt.plot(x, y)
plt.xticks(x)
plt.title("Simple line plot")
plt.legend(['y - random values'])
plt.show()
```

```python
# Bar:
plt.bar(x, y)
plt.xticks(x)
plt.title("Simple line plot")
plt.legend(['y - random values'])
plt.show()
```

```python
# Histogram:
x = np.random.randint(1,6, 100)
plt.hist(x, bins = 5, rwidth=0.5)
plt.xticks(x)
plt.title("Simple line plot")
plt.legend(['y - random values'])
plt.show()
```

```python
# Multiple Histograms: 
x1 = np.random.randint(1,6, 100)
x2 = np.random.randint(1,6, 100)
plt.hist([x1,x2], bins = 5)
plt.xticks(np.arange(1,6))
plt.title("Simple line plot")
plt.legend(['y2 - random values','y2 - random values',  ])
plt.show()
```

```python
# Pie chart, where the slices will be ordered and plotted counter-clockwise:
labels = 'Frogs', 'Hogs', 'Dogs', 'Logs'
sizes = [15, 30, 45, 10]
explode = (0, 0.1, 0.1, 0)  # only "explode" the 2nd slice (i.e. 'Hogs')

fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode,
        labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```

```python
# Area chart
plt.stackplot(np.arange(1, len(y1) + 1), y1, y2)
plt.xticks(np.arange(1, len(y1) + 1))
plt.legend(['y2 - random values','y2 - random values',  ])
plt.title("The stacked area chart")
plt.show()
```
